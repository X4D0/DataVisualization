### Soal 1 (15 poin)
Tampilkan visualisasi distribusi data untuk variabel "Glucose" dan "Insulin", lalu jelaskan hasil distribusinya.

#### Code

```python
# type your code here
# 1301184233
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('https://www.dl.dropboxusercontent.com/s/3i3gfr8pih4jwd4/diabetes.csv?dl=0') # Read csv
df = df.fillna(df.mean()) # Fill Missing Values with Mean in All Columns
#df.head() # Show Top 5 Data
df_GI = df[['Glucose','Insulin']] # Get Glucose and Insulin Only
#print(df_GI.head(5)) # Show Top 5 Glucose and Insulin
df_GI.plot(kind='box',title='Distribusi Glucose dan Insulin', figsize=(9,9)) # Membuat Boxplot untuk membandingkan distribusi angka Glucose dan Insulin terhadap Diabetes
print(df_GI.describe())
```

#### Penjelasan

Untuk menampilkan Visualisasi Distribusi Data pada Python maka langkah pertama yang perlu dilakukan adalah dengan *import library* Pandas yang didalam nya sudah terdapat fungsi untuk membaca Data dengan Dataframe sekaligus melakukan plotting terhadap data yang ada serta melakukan **Pre-Processing** Data. Untuk memberikan hasil yang baik maka perlu menghilangkan nilai null, pada kali ini Saya memilih untuk mengganti nilai Null dengan nilai Mean pada kolom tersebut. Lalu membuat dataframe baru yang berisikan dataframe df akan tetapi hanya kolom Glucose dan Insulin.

Selanjutnya untuk melakukan visualisasi distribusi data, digunakan **Boxplot** agar lebih mudah dan juga lebih spesifik. Pada hasil Boxplot yang ada dapat dilihat Garis Hitam diatas merupakan Nilai Maximum, Garis Hitam dibawah merupakan Nilai Minimum, Box bagian bawah adalah Nilai Q1, Box bagian atas adalah Nilai Q3, Garis Hijau merupakan nilai Q2 atau Mean, dan terakhir Lingkaran kecil merupakan Outliers yang ada.

Dari boxplot tersebut disimpulkan bahwa Dari Jumlah 768 Pasien didapatkan **Nilai Max dan Minimum** yang dimiliki oleh Glucose adalah 0 dan 199, sedangkan untuk **Insulin** adalah 0 dan 846. **Rata-rata nilai** Glucose yang dimiliki oleh semua Pasien adalah 120,89 dan Insulin yaitu 79,79. Dari gambar juga dapat dikatakan bahwa untuk Variable Glucose hanya terdapat sedikit data pecilan / Outliers sedangkan Insulin terdapat banyak Data Outliers nya.

### Soal 2 (25 poin)
Tampilkan visualisasi dari korelasi (i) antar variabel dan (ii) antara variabel dan target, lalu jelaskan hasil korelasinya.

#### Code

```python
# type your code here
df_kor1 = df.drop(['Outcome'], axis=1) # Create new dataframe contains all column exclude Outcome
df_kor2 = df
# Correlation 1
korelasi1 = df_kor1.corr()
# Correlation 2
korelasi2 = df_kor2.corr()[['Outcome']]
# Subplot
plt.figure(figsize=(12, 6))
plt.subplot(121)   #  subplot 1
plt.title('Korelasi 1 : Antar Variabel')
sn.heatmap(korelasi1, annot=True, fmt='.2f', square=True, cmap = 'Blues_r')

plt.subplot(122)   #  subplot 2
plt.title('Korelasi 2 : Variabel dengan Outcome')
sn.heatmap(korelasi2.sort_values(by='Outcome', ascending=False), vmin=-1, vmax=1, annot=True, fmt='.2f', square=True, cmap = 'Reds_r')

plt.show()
```

#### Penjelasan
#### Korelasi 1 
Untuk membandingkan korelasi 1, maka diperlukan dataframe yang berisikan seluruh variabel/kolom terkecuali Outcome. Setelah itu gunakan **fungsi corr()** untuk menghitung korelasi antar variabel, dan masukkan hasil fungsi tersebut kedalam Heatmap untuk divisualisasikan.

Hasil dari visualisasi tersebut didapat bahwa setiap variabel tentu saja akan memiliki nilai korelasi yang tinggi dalam hal ini ditanda kan dengan angka 1 dengan dirinya sendiri/variabel itu sendiri. Selain itu nilai korelasi yang tinggi juga terdapat antara **variabel Age dan Pregnancies** yang menandakan bahwa kedua variable tersebut saling berhubungan dan mempengaruhi dengan nilai 0.54 dari 1.00.

Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.

#### Code

```python
# type your code here
# Import Dataset
df_nasdaq = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/Nasdaq.csv')
df_hang = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/Hang_Seng.csv')
df_nikkei = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/Nikkei.csv')

# Drop the null Entries from Dataset
df_nasdaq = df_nasdaq.dropna(axis = 0)
df_hang = df_hang.dropna(axis = 0)
df_nikkei = df_nikkei.dropna(axis = 0)

# Change Date column type to Datetime64 and Set it to become Index
df_nasdaq['Date'] = pd.to_datetime(df_nasdaq['Date'])
df_hang['Date'] = pd.to_datetime(df_hang['Date'])
df_nikkei['Date'] = pd.to_datetime(df_nikkei['Date'])
df_nasdaq.index = df_nasdaq["Date"]
df_hang.index = df_hang["Date"]
df_nikkei.index = df_nikkei["Date"]

# Create a new Column 'Day_Perc_Change'
df_nasdaq["Day_Perc_Change"] = df_nasdaq["Adj Close"].pct_change()*100
df_hang["Day_Perc_Change"] = df_hang["Adj Close"].pct_change()*100
df_nikkei["Day_Perc_Change"] = df_nikkei["Adj Close"].pct_change()*100

# "Date" Column is duplicate, so we need to Drop one
df_nasdaq = df_nasdaq.drop(['Date'], axis=1)
df_hang = df_hang.drop(['Date'], axis=1)
df_nikkei = df_nikkei.drop(['Date'], axis=1)

# The first value in the ‘Day_Perc_Change’ column is NaN, so it needs to be drop
df_nasdaq = df_nasdaq.dropna(axis = 0)
df_hang = df_hang.dropna(axis = 0)
df_nikkei = df_nikkei.dropna(axis = 0)

# Trend Analysis
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"

df_nasdaq["Trend"]= df_nasdaq["Day_Perc_Change"].apply(lambda x:trend(x))
df_hang["Trend"]= df_hang["Day_Perc_Change"].apply(lambda x:trend(x))
df_nikkei["Trend"]= df_nikkei["Day_Perc_Change"].apply(lambda x:trend(x))

# Only take 2019 Stock Market
df_nasdaq = df_nasdaq['2019-01-01' : '2019-12-31']
df_hang = df_hang['2019-01-01' : '2019-12-31']
df_nikkei = df_nikkei['2019-01-01' : '2019-12-31']

# Visualize with Pie Chart
plt.figure(figsize=(12, 12))

plt.subplot(131)   #  subplot 1
plt.title('Trend NASDAQ', y=1.5)
pie_nasdaq = df_nasdaq.groupby('Trend')
pie_label = sorted([i for i in df_nasdaq.loc[:, 'Trend'].unique()])
plt.pie(pie_nasdaq['Trend'].count(), shadow=True, labels=None, autopct = '%1.1f%%', radius = 2, pctdistance=1.12)
plt.legend(labels=pie_label, loc='upper left', bbox_to_anchor = (1.5, 1)) 

plt.subplot(132) # subplot 2
plt.title('Trend HANG SENG', y=1.5)
pie_hang = df_hang.groupby('Trend')
pie_label = sorted([i for i in df_hang.loc[:, 'Trend'].unique()])
plt.pie(pie_hang['Trend'].count(), shadow=True, labels=None, autopct = '%1.1f%%', radius = 2, pctdistance=1.12)
plt.legend(labels=pie_label, loc='upper left', bbox_to_anchor = (1.5, 1)) 

plt.subplot(133) # subplot 3
plt.title('Trend NIKKEI', y=1.5)
pie_nikkei = df_nikkei.groupby('Trend')
pie_label = sorted([i for i in df_nikkei.loc[:, 'Trend'].unique()])
plt.pie(pie_nikkei['Trend'].count(), shadow=True, labels=None, autopct = '%1.1f%%', radius = 2, pctdistance=1.12)
plt.legend(labels=pie_label, loc='upper left', bbox_to_anchor = (1.5, 1)) 

plt.subplots_adjust(left=2, bottom=8, right=3.3, top=9, wspace=2, hspace=2)
plt.show()
```

#### Penjelasan 
Berdasarkan hasil visualisasi diatas, dapat kita simpulkan bahwa Market milik **NASDAQ lebih baik** dibandingan yang lainnya. Hal ini dapat kita lihat bahwa sebesar **17.1%** dari periode waktu di Tahun 2019 stock market NASDAQ memiliki peningkatan harga/kenaikan positif. Sedangkan HANG SENG hanya sebesar 13.0% dan NIKKEI 11.6%. Lalu untuk stock market yang memiliki penurunan harga saham terbanyak ada pada HANG SENG dimana sebesar 12.2% dari periode waktu di Tahun 2019 terjadi penurunan harga/negatif. Selanjutnya pada Market NIKKEI dapat terbilang hampir stabil dimana sekitar 52.7% dari periode waktunya hampir tidak ada perubahan harga (Naik atau Turun).

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# type your code here
volatility = pd.DataFrame()
volatility['NASDAQ'] = df_nasdaq["Day_Perc_Change"]
volatility['HANGSENG'] = df_hang["Day_Perc_Change"]
volatility['NIKKEI'] = df_nikkei["Day_Perc_Change"]

# Since the volatility Dataframe has Null, we need to drop the row
volatility.dropna(inplace=True, axis=0)

plotvolatil = volatility[['NASDAQ', 'HANGSENG', 'NIKKEI']].rolling(7).std()*np.sqrt(7)
plotvolatil.plot(figsize=(15,6))
```

#### Penjelasan 
Sebagaimana hasil visualisasi diatas, dapat dilihat bahwa Saham milik NIKKEI terbilang cukup stabil dimana volatilitas nya terdapat di kisaran 1-3 meskipun titik tertingginya hampir menyentuh 5 dan sempat mencapai titik terendah nya pada bulan November. Hal serupa juga terjadi pada Saham HANG SENG dan berada pada titik tertinggi diakhir Tahun 2019 dibandingkan dengan Saham yang lain. Lalu ditahun 2019 ini volatil tertinggi didapatkan oleh Saham NASDAQ di sekitar bulan September.
